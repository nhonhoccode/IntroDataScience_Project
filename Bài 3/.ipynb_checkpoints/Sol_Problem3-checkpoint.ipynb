### 3.1.1 - Bạn hãy viết code cào dữ liệu từ trang web trên, lưu kết quả vào 1 file tương ứng (kq.txt) và vẽ sơ đồ mô tả ngắn gọn về cấu trúc của trang Web trên?

```python
def scrape_and_save(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        with open('kq.txt', 'a', encoding='utf-8') as file:
            file.write(response.text)
            file.write('\n\n') 
        print(f"Nội dung HTML của trang {url} đã được cào và lưu vào file kq.txt!")
        
        # Tìm liên kết của trang tiếp theo
        next_page_link = soup.find('li', class_='next')
        if next_page_link:
            next_url = next_page_link.find('a')['href']
            next_full_url = urljoin(url, next_url)
            scrape_and_save(next_full_url)
    else:
        print(f"Không thể truy cập trang {url}.")

url = 'https://quotes.toscrape.com/'

scrape_and_save(url)
```

### c. Hãy viết hàm tacgiaLink() để lấy nội dung của mỗi tác giả. Với mỗi tác giả in ra màn hình các nội dung:
- Tên tác giả
- Đường link của tác giả
- Ngày tháng năm sinh
- Và câu nói nổi tiếng của tác giả

```python
# Hàm để lấy tên tác giả
def get_author_name(quote):
    author_tag = quote.find('small', class_='author')
    if author_tag:
        return author_tag.text(strip=True)
    else:
        return None

# Hàm để lấy đường link của tác giả
def get_author_link(quote):
    author_tag = quote.find('small', class_='author')
    if author_tag:
        author_link = author_tag.find_parent('div', class_='quote').find('a')['href']
        return 'https://quotes.toscrape.com' + author_link
    else:
        return None

# Hàm để lấy ngày tháng năm sinh của tác giả
def get_author_birth_date(author_link):
    response = requests.get(author_link)
    if response.status_code == 200:
        author_soup = BeautifulSoup(response.content, 'html.parser')
        birth_date_tag = author_soup.find('span', class_='author-born-date')
        if birth_date_tag:
            return birth_date_tag.get_text(strip=True)
        else:
            return None
    else:
        return None

def get_author_quotes(author_name):
    page_number = 1
    quotes = []

    while True:
        # Định dạng URL của trang kế tiếp
        url = f'https://quotes.toscrape.com/page/{page_number}/'
        
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            # Tìm tất cả các trích dẫn của tác giả
            for quote in soup.find_all('div', class_='quote'):
                author = quote.find('small', class_='author')
                if author and author.get_text(strip=True) == author_name:
                    quote_text = quote.find('span', class_='text')
                    if quote_text:
                        quotes.append(quote_text.get_text(strip=True))
            # Kiểm tra xem còn trang kế tiếp không
            next_page = soup.find('li', class_='next')
            if next_page:
                page_number += 1
            else:
                break
        else:
            print(f"Không thể truy cập trang {url}")
            break

    return quotes

# Hàm để thu thập thông tin của mỗi tác giả
def collect_author_info():
    author_links = set()  # Sử dụng set để loại bỏ các đường link trùng lặp
    
    # Đọc nội dung từ file kq.txt
    with open('kq.txt', 'r', encoding='utf-8') as file:
        html_content = file.read()

    # Phân tích nội dung HTML bằng BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Tìm tất cả các trích dẫn và lấy đường link của tác giả
    quotes = soup.find_all('div', class_='quote')
    for quote in quotes:
        author_link = get_author_link(quote)
        if author_link:
            author_links.add(author_link)  # Thêm đường link vào set để loại bỏ các link trùng lặp

    return list(author_links)  # Chuyển set thành list để dễ dàng duyệt qua và sử dụng


# Hàm tổng hợp thông tin của mỗi tác giả
def tacgiaLink():
    author_links = collect_author_info()

    # Lặp qua từng đường link của tác giả để lấy thông tin và in ra màn hìnhxS
    for author_link in author_links:
        # Lấy tên tác giả
        response = requests.get(author_link)
        if response.status_code == 200:
            author_soup = BeautifulSoup(response.content, 'html.parser')
            author_name = author_soup.find('h3', class_='author-title').text
            print("---------------------------------")
            print("Tên tác giả:", author_name)
            # In đường link của tác giả
            print("Đường link của tác giả:", author_link)
            # Lấy ngày tháng năm sinh của tác giả
            birth_date = get_author_birth_date(author_link)
            if birth_date:
                print("Ngày tháng năm sinh:", birth_date)
            else:
                print("Không tìm thấy thông tin về ngày tháng năm sinh của tác giả.")
            # Lấy câu nói nổi tiếng của tác giả
            if(author_name == "Alexandre Dumas-fils"):
                famous_quote = get_author_quotes("Alexandre Dumas fils")
            else:
                famous_quote = get_author_quotes(author_name)
            if famous_quote:
                print("Câu nói nổi tiếng của tác giả:", famous_quote)
            else:
                print("Không tìm thấy câu nói nổi tiếng của tác giả.")
        else:
            print("Không thể truy cập đường link của tác giả.")

tacgiaLink()
```

### d. Hãy lưu kết quả ở câu c vào file Quote.csv tương ứng, với mỗi tác giả là 1 dòng dữ liệu. Bạn được yêu cầu thu thập ít nhất 40 câu nói nổi tiếng từ trang web trên một cách tự động theo code của các ý trên

```python
# Hàm để thu thập thông tin của mỗi tác giả
def collect_author_info():
    author_links = set()  
    
    # Đọc nội dung từ file kq.txt
    with open('kq.txt', 'r', encoding='utf-8') as file:
        html_content = file.read()

    soup = BeautifulSoup(html_content, 'html.parser')

    # Tìm tất cả các trích dẫn và lấy đường link của tác giả
    quotes = soup.find_all('div', class_='quote')
    for quote in quotes:
        author_link = get_author_link(quote)
        if author_link:
            author_links.add(author_link)  # Thêm đường link vào set để loại bỏ các link trùng lặp

    return list(author_links)  # Chuyển set thành list để dễ dàng duyệt qua và sử dụng

# Hàm để ghi thông tin của các tác giả vào file CSV
def write_to_csv(author_info):
    with open('Quote.csv', 'w', newline='', encoding='utf-8') as csvfile:
        fieldnames = ['Author Name', 'Link', 'Birth', 'Quotes', 'Number Quotes']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        
        # Ghi tiêu đề cho các cột trong file CSV
        writer.writeheader()
        
        # Lặp qua từng thông tin của tác giả và ghi vào file CSV
        for info in author_info:
            writer.writerow(info)

# Hàm để tổng hợp thông tin của mỗi tác giả
def tacgiaLink():
    author_links = collect_author_info()
    author_info = []

    for author_link in author_links:
        response = requests.get(author_link)
        if response.status_code == 200:
            author_soup = BeautifulSoup(response.content, 'html.parser')
            author_name = author_soup.find('h3', class_='author-title').get_text(strip=True)
            
            birth_date = get_author_birth_date(author_link)
            
            if(author_name == "Alexandre Dumas-fils"):
                famous_quote = get_author_quotes("Alexandre Dumas fils")
            else:
                famous_quote = get_author_quotes(author_name)
            
            author_info.append({'Author Name': author_name, 'Link': author_link, 'Birth': birth_date, 'Quotes': famous_quote, 'Number Quotes': len(famous_quote)})
    
    write_to_csv(author_info)  # Gọi hàm để ghi thông tin vào file CSV

# Gọi hàm để tổng hợp thông tin của các tác giả và ghi vào file CSV
tacgiaLink()
```

### Bạn hãy thêm vào Trường Tuoi (Tuổi) và đề xuất cách điền tuổi của các tác giả?

Như đã nói ở trên em sẽ dùng phương pháp 1 trong 2 phương pháp em đã nêu ra đó là sử dụng google search và selenium để crawl thêm được ngày mất của tác giả và từ đó tính được tuổi tác nếu còn sống thì để Alive

```python
Deathday = []
driver = webdriver.Chrome()
googlesearch = 'https://www.google.com/search?q='
driver.get('https://google.com')
for idx, name in enumerate(df['Author Name']):
    driver.get(googlesearch + name)
    sleep(2)
    try:
        deathday = driver.find_element(By.CSS_SELECTOR, 'div:nth-child(3) > div > div > div > span:nth-child(2) > span')
        author_death = deathday.text.split(', ')[1]
        if int(author_death) - int(df['Birth'][idx][-4:]) > 200 or int(author_death) - int(df['Birth'][idx][-4:]) <= 0:
            assert(False)
        print(f"{name}: ({df['Birth'][idx][-4:]}:{author_death})")
        Deathday.append(author_death)
    except:
        Deathday.append(None)
        print(f"{name}: ({df['Birth'][idx][-4:]}: ALive)")
    sleep(2)
```

```python
df['Year of Death'] = Deathday
df
```

```python
df['Year of Birth'] = df['Birth'].apply(lambda x: int(x[-4:]))
df
```

```python
current_year = datetime.now().year
list_age = []
for idx in range(len(df['Year of Death'])):
    deathday = df['Year of Death'][idx]
    if deathday == None:
        list_age.append(current_year - df['Year of Birth'][idx])
    else:
        list_age.append(int(deathday) - df['Year of Birth'][idx])
list_age
```

```python
# Hàm tách quote
def solve_quotes(quotes):
    result = []
    for quote in quotes.replace('“', '”').split('”'):
        if len(quote) > 5:
            result.append(quote)
    return result
```

```python
Names_df_quote = []
Births_df_quote = []
Deaths_df_quote = []
Ages_df_quote = []
Quotes_df_quote = []
index = 0
for id_author in range(len(df['Author Name'])):
    list_quotes = solve_quotes(df['Quotes'][id_author])
    for id_quote in range(len(list_quotes)):
        Names_df_quote.append(df['Author Name'][id_author])
        Births_df_quote.append(df['Year of Birth'][id_author])
        Deaths_df_quote.append(df['Year of Death'][id_author])
        Ages_df_quote.append(df['Age'][id_author])
        
        Quotes_df_quote.append(list_quotes[id_quote])

df_quotes['Author Name'] = Names_df_quote
df_quotes['Year of Birth'] = Births_df_quote
df_quotes['Age'] = Ages_df_quote
df_quotes['Year of Death'] = Deaths_df_quote
df_quotes['Quotes'] = Quotes_df_quote
df_quotes
```

```python
def Age_Top(df_vis, x_col, y_col, num_top, check_sort, color, x_label, y_label, title_vis):
    topnum = df_vis.sort_values(by = x_col, ascending = check_sort).head(num_top)
    sns.set_style("whitegrid")
    plt.figure(figsize=(7, 5))
    plt.barh(topnum[y_col][::-1], topnum[x_col][::-1], color=color)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title_vis)
    for i, v in enumerate(topnum[x_col][::-1]):
        plt.text(v + 1, i, str(v), color=color, fontweight='bold', fontsize = 12)
    plt.show()
```

### Thống kê về các từ được sử dụng trong các câu nói

```python
def get_word(x):
    check = [',', '.', '-', "'", '"', '?', '!', ':']
    ans = []
    for word in x.split():
        if word not in check:
            word = word.replace(',', '')
            word = word.replace('.', '')
            word = word.replace('!', '')
            word = word.replace("'", '')
            word = word.replace('"', '')
            word = word.replace('?', '')
            word = word.replace(':', '')
            word = word.lower()
            ans.append(word)
    return ans
full_word = []
for quote in df_quotes['Quotes']:
    full_word += get_word(quote)
full_word
```

```python
set_word = list(set(full_word))
num_word = []
for word in set_word:
    num_word.append(full_word.count(word))
num_word
```

```python
df_word = pd.DataFrame(columns = ['Word', 'Count'])
df_word['Word'] = set_word
df_word['Count'] = num_word
df_word.head(20)
```

```python
def prep_text(df, col):
    """
    Thay thế các non-alphanumeric thành khoảng cách, còn lại sẽ chuyển thành chữ thường
    Thay thế khoảng trắng liên tiếp thành một khoảng trắng, và loại bỏ khoảng trắng ở đầu và cuối
    """
    return df[col].apply(lambda x: re.sub('[^A-Za-z0-9]+', ' ', x.lower()))\
                  .apply(lambda x: re.sub('\s+', ' ', x).strip())
```

```python
class Model:
    def __init__(self, model, name):
        self.model = model
        self.name = name
        self.is_trained = False
        self.predicted = False
        self.ans = ""
        
    def fit(self, Xtrain, Xtest, ytrain, ytest):
        self.model.fit(Xtrain, ytrain)
        ypredict = self.model.predict(Xtest)
        self.acc = accuracy_score(ytest, ypredict)
        self.f1_score = f1_score(ytest, ypredict, average='macro')
        self.is_trained = True
    def Prediction(self, your_predict):
        self.predicted = True
        return self.model.predict(your_predict)
    
    def __repr__(self):
        if self.is_trained:
            return f'{self.name} has Accuracy Score: {round(self.acc, 4)} \n {self.name} has F1 score: {round(self.f1_score, 4)}'
        return f'{self.name} is not trained yet'
```
